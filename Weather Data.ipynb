```python
43. Create a new column categorizing 'Visibility' into 'Low', 'Moderate', and 'High'.
```

```python
wea.head(1)
```

```python
bins = [0, 5, 15, wea['Visibility_km'].max()]  # Adjust if needed
```

```python
Here we are defining the cutoff points for each category:

0 to 5 km → Low

>5 to 15 km → Moderate

>15 to max visibility in the dataset → High

The df['Visibility_km'].max() part ensures the highest value is captured properly, no matter what your data looks like.

These are the labels you want to assign to each bin.
```

```python
labels = ['Low', 'Moderate', 'High']
```

```python
These are the labels you want to assign to each bin.
```

```python
wea['Visibility_Level'] = pd.cut(wea['Visibility_km'], bins=bins, labels=labels, include_lowest=True)
```

Here’s what’s happening:

df['Visibility_km']: the column we’re categorizing.

bins=bins: we apply the bins defined earlier.

labels=labels: each bin gets a label.

include_lowest=True: makes sure the lowest value is included in the first bin (e.g., if visibility is 0 km, it's counted as Low).
